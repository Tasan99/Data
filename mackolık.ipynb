```python
team_data_list = []


teams = soup.find_all("tr", {'class': "alt1 row"})
for team in teams:
    team_info = team.text.strip().split('\n')
    team_data = {
        "Team": team_info[0],
        "League": team_info[1],
        "Game": team_info[2],
        "Goal/Game": team_info[3],
        "Shoot/Game": team_info[4],
        "Isabetli/Shoot": team_info[5],
        "TSO": team_info[6],
        "Pass": team_info[7],
        "B.Pas/M": team_info[8],
        "IMK": team_info[9],
        "HT": team_info[10],
        "Kor/M":team_info[11]
    }
    team_data_list.append(team_data)

print(team_data_list)
# Assuming team_data_list is your list of dictionaries

# Find the index of Bayern in the list
bayern_index = None
for i, team_data in enumerate(team_data_list):
    if team_data["Team"] == "Bayern Münih":
        bayern_index = i
        break

# Print specific fields for Bayern if found
if bayern_index is not None:
    bayern_team = team_data_list[bayern_index]
    print(bayern_team["Team"], bayern_team["League"], bayern_team["Game"], bayern_team["Goal/Game"], bayern_team["Shoot/Game"], bayern_team["Isabetli/Shoot"])
    if float(bayern_team["Goal/Game"].replace(',','.'))>=1.5:
        print("Bayern 1,5 OYNA")
else:
    print("Bayern not found in the list.")


```

```python
from bs4 import BeautifulSoup
import requests
import time

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
}
url = "https://www.mackolik.com/canli-sonuclar"
response = requests.get(url, headers=headers)


def get_live_match_data():
   
    
    while True:
        try:
            # Web sayfasını getir
            response = requests.get(url)
            response.raise_for_status()  # HTTP hatası varsa hemen hata fırlat
            soup = BeautifulSoup(response.text, 'html.parser')

            # Extract relevant information
            home_team_name = soup.select_one('.match-row__team-name--home .match-row__team-name-text').text.strip()
            away_team_name = soup.select_one('.match-row__team-name--away .match-row__team-name-text').text.strip()
            score_home = soup.select_one('.match-row__score-home').text.strip()
            score_away = soup.select_one('.match-row__score-away').text.strip()
            match_link = soup.select_one('.match-row__score')['href']

            # Print the extracted information
            print(f"Home Team: {home_team_name}")
            print(f"Away Team: {away_team_name}")
            print(f"Score: {score_home} - {score_away}")
            print(f"Match Link: {match_link}")
            
            time.sleep(60)  # 60 saniye bekleyerek tekrar kontrol et
        except Exception as e:
            print(f"Hata oluştu: {e}")
            time.sleep(60)  # Hata durumunda 60 saniye bekleyerek tekrar deneyin

if __name__ == "__main__":
    get_live_match_data()
```
